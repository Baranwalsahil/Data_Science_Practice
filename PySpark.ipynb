```python
df = spark.createDataFrame([
                  ['red', 'banana', 1, 10], ['blue', 'banana', 2, 20], ['red', 'carrot', 3, 30],
    ['blue', 'grape', 4, 40], ['red', 'carrot', 5, 50], ['black', 'carrot', 6, 60],
    ['red', 'banana', 7, 70], ['red', 'grape', 8, 80]], schema=['color', 'fruit', 'v1', 'v2'])
df
```

```python
df.show()
```

```python
df.groupby('color').avg().show()
```

```python
def plus_mean(pandas_df):
  return pandas_df.assign(v1 = pandas_df.v1 - pandas_df.v1.mean())
df.groupby('color').applyInPandas(plus_mean,schema=df.schema).show()
```

```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
dept = spark.sparkContext.parallelize([
       ("Finance",10),("Marketing",20),("Sales",30),("IT",40)
])
dept.collect()
```

```python
from pyspark.sql.types import StructType, StructField,IntegerType, FloatType, StringType,TimestampType,DateType

data = [('James','','Smith','2017-08-02 16:16:14','M',3000),
   ('Michael','Rose','Rose','1968-03-05 04:53:05','M',4000),
   ('Robert','','Williams','1978-09-05 04:53:05','M',4000),
   ('Maria','Anne','Jones','1967-12-01 04:53:05','F',4000),
   ('Jen','Mary','Brown','1980-02-17 04:53:05','F',-1)]
```

```python
columns = StructType([
    StructField("firstname", StringType(), True),
    StructField("middlename", StringType(), True),
    StructField("lastname", StringType(), True),
    StructField("dob", StringType(), True),
    StructField("gender", StringType(), True),
    StructField("salary", IntegerType(), True)])
```

```python
df = spark.createDataFrame(data,columns)
df.show()
df.printSchema()
```

```python
deptCol = ["Dept","Salary"]
df1 = spark.createDataFrame(dept,schema=deptCol)
df1.show()
df1.printSchema()
```
